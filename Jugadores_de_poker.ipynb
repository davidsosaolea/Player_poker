```python
import pandas as pd
from google.oauth2.credentials import Credentials
from googleapiclient.discovery import build
from google.oauth2 import service_account

# DataFrame con enlaces y fechas

# Función para procesar cada enlace
def process_spreadsheet(enlace, fecha):
    SCOPES = ['https://www.googleapis.com/auth/spreadsheets']
    KEY = '/content/key.json'

    # Extraer el ID del documento del enlace
    SPREADSHEET_ID = enlace.split('/')[5]

    # Credenciales y servicio de Google Sheets
    creds = service_account.Credentials.from_service_account_file(KEY, scopes=SCOPES)
    service = build('sheets', 'v4', credentials=creds)
    sheet = service.spreadsheets()

    # Llamada a la API para obtener los datos
    result = sheet.values().get(spreadsheetId=SPREADSHEET_ID, range='Final results!A4:W').execute()
    values = result.get('values', [])

    # Convertir los datos a un DataFrame de pandas
    df = pd.DataFrame(values[1:], columns=values[0])

    # Renombrar las columnas 'Profit' y 'EV' a 'Profit$' y 'EV$' si existen
    df.rename(columns={'Profit': 'Profit$', 'EV': 'EV$'}, inplace=True)

    df = df[['Nickname', 'Stake', 'Distance', 'EV Chips', 'ChEV', 'Profit$', 'EV$']]

    # Eliminar filas donde 'Stake' está vacío o es None
    df = df[df['Stake'].notna() & (df['Stake'] != '')]

    # Cambiar las columnas a tipo numérico entero
    columns = ['Distance', 'EV Chips']
    df[columns] = df[columns].apply(pd.to_numeric, errors='coerce').fillna(0).astype(int)

    # Cambiar las columnas a tipo numérico flotante
    columns = ['ChEV', 'Profit$', 'EV$']
    df[columns] = df[columns].apply(pd.to_numeric, errors='coerce').fillna(0.0).astype(float)

    df.loc[:, 'Diff'] = abs(df['EV$'] - df['Profit$'])

    # Eliminar filas donde la columna 'Distance' es igual a 0
    df = df[df['Distance'] != 0]

    # Utilizar str.extract con expresiones regulares para separar las partes
    df[['BI', 'Divisa', 'Descripcion']] = df['Stake'].str.extract(r'(\d+)(€|\$)?\s*(.*)')

    # Colocar '$' en 'Divisa' si 'BI' es NaN en el original (antes de fillna)
    df.loc[df['Divisa'].isna(), 'Divisa'] = '$'

    # Convertir 'BI' a tipo numérico
    df['BI'] = pd.to_numeric(df['BI'])

    # Actualizar la columna 'modalidad' basado en la condición de la columna 'Descripcion'
    df['modalidad'] = df['Descripcion'].apply(lambda x: 'HU' if 'HUHT' in x else 'SPIN')

    # Lista de palabras clave para identificar TURBO
    keywords = ['HUHT', 'nitro', 'flash', 'GG', 'ultra']

    # Función para verificar si alguna palabra clave está en la descripción
    def check_turbo(description):
        return 'TURBO' if any(keyword in description for keyword in keywords) else 'NORMAL'

    # Actualizar la columna 'Velocidad' basado en la condición de la columna 'Descripcion'
    df['Velocidad'] = df['Descripcion'].apply(check_turbo)

    # Asignar valores a la columna 'sala' según el contenido de 'Stake'
    df.loc[df['Descripcion'].str.contains('wx', na=False), 'Sala'] = 'Winamax'
    df.loc[df['Descripcion'].str.contains('nitro', na=False), 'Sala'] = 'Winamax'

    df.loc[df['Descripcion'].str.contains('GG', na=False), 'Sala'] = 'GG Poker'

    df.loc[df['Descripcion'] == '(ipoker)', 'Sala'] = 'Ipoker'
    df.loc[df['Descripcion'] == '(ipoker HUHT)', 'Sala'] = 'Ipoker'

    df.loc[df['Descripcion'] == '(ipoker.it)', 'Sala'] = 'Ipoker IT'
    df.loc[df['Descripcion'] == '(ipoker.es)', 'Sala'] = 'Ipoker ES'
    df.loc[df['Descripcion'] == '(ipoker.fr)', 'Sala'] = 'Ipoker FR'

    df.loc[df['Descripcion'] == '(.it flash) new', 'Sala'] = 'Pokerstars IT'
    df.loc[df['Descripcion'] == '(.it) new', 'Sala'] = 'Pokerstars IT'
    df.loc[df['Descripcion'] == '(.it)', 'Sala'] = 'Pokerstars IT'
    df.loc[df['Descripcion'] == '(.it flash)', 'Sala'] = 'Pokerstars IT'

    df.loc[df['Descripcion'] == '(party-es)', 'Sala'] = 'Party Poker ES'
    df.loc[df['Descripcion'] == '(ultra-es)', 'Sala'] = 'Party Poker ES'
    df.loc[df['Descripcion'] == '(party-es HUHT)', 'Sala'] = 'Party Poker ES'

    df.loc[df['Descripcion'] == '(pa HUHT)', 'Sala'] = 'Party Poker'
    df.loc[df['Descripcion'] == 'pa', 'Sala'] = 'Party Poker'
    df.loc[df['Descripcion'] == '(party-ca)', 'Sala'] = 'Party Poker'
    df.loc[df['Descripcion'] == '(party-overdrive)', 'Sala'] = 'Party Poker'
    df.loc[df['Descripcion'] == '(party)', 'Sala'] = 'Party Poker'
    df.loc[df['Descripcion'] == '(ultra)', 'Sala'] = 'Party Poker'
    df.loc[df['Descripcion'] == '(party HUHT)', 'Sala'] = 'Party Poker'


    df.loc[(df['Divisa'] == '$') & (df['Descripcion'] == '(HUHT) new'), 'Sala'] = 'Pokerstars'
    df.loc[(df['Divisa'] == '$') & (df['Descripcion'] == '(HUHT)'), 'Sala'] = 'Pokerstars'
    df.loc[(df['Divisa'] == '$') & (df['Descripcion'] == ''), 'Sala'] = 'Pokerstars'
    df.loc[(df['Divisa'] == '$') & (df['Descripcion'] == 'new'), 'Sala'] = 'Pokerstars'
    df.loc[(df['Divisa'] == '$') & (df['Descripcion'] == '(1million) new'), 'Sala'] = 'Pokerstars'
    df.loc[(df['Divisa'] == '$') & (df['Descripcion'] == '(1million)'), 'Sala'] = 'Pokerstars'
    df.loc[(df['Divisa'] == '$') & (df['Descripcion'] == '(flash) new'), 'Sala'] = 'Pokerstars'
    df.loc[(df['Divisa'] == '$') & (df['Descripcion'] == '(flash 1million) new'), 'Sala'] = 'Pokerstars'
    df.loc[(df['Divisa'] == '$') & (df['Descripcion'] == '(1 mil flash) new'), 'Sala'] = 'Pokerstars'
    df.loc[(df['Divisa'] == '$') & (df['Descripcion'] == '(1 mil flash)'), 'Sala'] = 'Pokerstars'
    df.loc[(df['Divisa'] == '$') & (df['Descripcion'] == '(flash)'), 'Sala'] = 'Pokerstars'
    df.loc[(df['Divisa'] == '$') & (df['Descripcion'] == '(flash 1million)'), 'Sala'] = 'Pokerstars'
    df.loc[(df['Divisa'] == '$') & (df['Descripcion'] == '(flash) 50%'), 'Sala'] = 'Pokerstars'
    df.loc[(df['Divisa'] == '$') & (df['Descripcion'] == '.5$ (HUHT)'), 'Sala'] = 'Pokerstars'
    df.loc[(df['Divisa'] == '$') & (df['Descripcion'] == '50%'), 'Sala'] = 'Pokerstars'


    df.loc[(df['Divisa'] == '€') & (df['Descripcion'] == '(HUHT) new'), 'Sala'] = 'Pokerstars ES'
    df.loc[(df['Divisa'] == '€') & (df['Descripcion'] == '(HUHT)'), 'Sala'] = 'Pokerstars ES'
    df.loc[(df['Divisa'] == '€') & (df['Descripcion'] == ''), 'Sala'] = 'Pokerstars ES'
    df.loc[(df['Divisa'] == '€') & (df['Descripcion'] == 'new'), 'Sala'] = 'Pokerstars ES'
    df.loc[(df['Divisa'] == '€') & (df['Descripcion'] == '(1million) new'), 'Sala'] = 'Pokerstars ES'
    df.loc[(df['Divisa'] == '€') & (df['Descripcion'] == '(1million)'), 'Sala'] = 'Pokerstars ES'
    df.loc[(df['Divisa'] == '€') & (df['Descripcion'] == '(flash) new'), 'Sala'] = 'Pokerstars ES'
    df.loc[(df['Divisa'] == '€') & (df['Descripcion'] == '(flash 1million) new'), 'Sala'] = 'Pokerstars ES'
    df.loc[(df['Divisa'] == '€') & (df['Descripcion'] == '(1 mil flash) new'), 'Sala'] = 'Pokerstars ES'
    df.loc[(df['Divisa'] == '€') & (df['Descripcion'] == '(1 mil flash)'), 'Sala'] = 'Pokerstars ES'
    df.loc[(df['Divisa'] == '€') & (df['Descripcion'] == '(flash)'), 'Sala'] = 'Pokerstars ES'
    df.loc[(df['Divisa'] == '€') & (df['Descripcion'] == '(flash 1million)'), 'Sala'] = 'Pokerstars ES'
    df.loc[(df['Divisa'] == '€') & (df['Descripcion'] == '(flash) 50%'), 'Sala'] = 'Pokerstars ES'
    df.loc[(df['Divisa'] == '€') & (df['Descripcion'] == '.5$ (HUHT)'), 'Sala'] = 'Pokerstars'
    df.loc[(df['Divisa'] == '€') & (df['Descripcion'] == '50%'), 'Sala'] = 'Pokerstars'

    # Agregar la fecha a los datos procesados
    df['FECHA'] = fecha

    return df

# Lista para almacenar los DataFrames procesados
dfs = []

# Iterar sobre cada fila del DataFrame SP y procesar cada enlace
for index, row in SP.iterrows():
    fecha = row['FECHA']
    enlace = row['ENLACE']
    df_processed = process_spreadsheet(enlace, fecha)
    dfs.append(df_processed)

# Concatenar todos los DataFrames procesados
result_df = pd.concat(dfs, ignore_index=True)

# Mostrar el DataFrame resultante
# print(result_df)
```
